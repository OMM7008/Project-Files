```python
import pandas as pd
import numpy as np
import seaborn as sns

```

```python
#Importing the dataset

df = pd.read_csv(r'D:\Microsoft VS Code\Data Analysis Using Python\my_venv\Diwali Sales Data.csv',encoding='unicode_escape')

print('The dataset is given by : \n')
df.head(20)
```

```python
print('The shape of the dataset before any data-cleaning: ',df.shape)

print("\n Basic info of the dataset is given by: \n")
df.info()
```

```python
num_cols = []
cat_cols = []

for cols in df.columns:
    if(df[cols].dtype == 'O'):
        cat_cols.append(cols)
    else:
        num_cols.append(cols)
print("The numerical columns are given by: \n", num_cols)
print("The categorical columns are given by: \n", cat_cols)
```

```python
#Checking for null-value columns
df.isnull().sum()
```

```python
#Dropping the null_columns from the dataset

df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#Checking the current shape 
df.shape
```

```python
#dropping the null_values from the dataset 

df.dropna(inplace=True)
print('After Dropping the null_values: \n')
df.shape
```

```python
#Changing its datatpye to "int"

df['Amount'] = df['Amount'].astype('int')
df['Amount'].dtype 
```

```python
df.head(15)
```

```python
#Descriptive Analysis of the Amount_Column

print("Amount_Column \n")
df['Amount'].describe()
```

EDA (Exploratory Data Analysis)

```python
#Gender Division plot

plt_1 = sns.countplot(x = 'Gender', data=df)

for bars in plt_1.containers:
    plt_1.bar_label(bars)
```

```python
#Sales division by gender

df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
Sale_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt_2 = sns.barplot(x = 'Gender',y = 'Amount',data=Sale_gen,)

for bars in plt_2.containers:
    plt_2.bar_label(bars)
    
```

Hence from the above plots we got an idea that females are ones who purchase items for diwali celebration as compared to males. 

```python
#Division by AGE category

plt_3 = sns.countplot(x = 'Age Group',hue='Gender', data=df)

for bars in plt_3.containers:
    plt_3.bar_label(bars)
```

```python
# Plot Between total_amount vs Age_group
Sale_Age_Group = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt_4 = sns.barplot(x = 'Age Group', y = 'Amount', data=Sale_Age_Group)

for bars in plt_4.containers:
    plt_4.bar_label(bars)
```

Hence Women belonging from 26-35 age group do the maximum purchase.


```python
#Top 10 states on the basis of maximum no of orders

Sales_states_order = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize': (15,5)})
plt_5 = sns.barplot(x = 'State', y = 'Orders', data=Sales_states_order)

for bars in plt_5.containers:
    plt_5.bar_label(bars)
```

```python
#Top 10 states with with maximum sales amount

Sales_states_amount = df.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
plt_6 = sns.barplot(x = 'State', y= 'Amount', data=Sales_states_amount)

for bars in plt_6.containers:
    plt_6.bar_label(bars)
```

Hence UP is the state with the maximum sales with the highest no of orders.

```python
#Division by Marital Status

plt_7 = sns.countplot(x = 'Marital_Status', data=df)
sns.set(rc = {'figure.figsize':(4,5)})
for bars in plt_7.containers:
    plt_7.bar_label(bars)

```

```python
#Top 5 occupation with maximum sales amount on the basis of gender

Sales_Occupation_Amount = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(5)
sns.set(rc = {'figure.figsize':(10,5)})
plt_8 = sns.barplot(x = 'Occupation', y = 'Amount', data=Sales_Occupation_Amount)

for bars in plt_8.containers:
    plt_8.bar_label(bars)
```

Hence from the above graph we got to know that the maximum sales was made from the IT sector followed by Health sector.

```python
#Sales by top 5 Product_Category

Sales_Product_category = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
plt_9 = sns.barplot(x = 'Amount', y = 'Product_Category', data=Sales_Product_category)
sns.set(rc={'figure.figsize':(10,5)})

for bars in plt_9.containers:
    plt_9.bar_label(bars)
```

```python
#Top 10 most selling products

Sales_Product_ID = df.groupby(['Product_ID'], as_index= False)['Orders'].sum().sort_values(by = 'Orders', ascending= False).head(10)
plt_10 = sns.barplot(y = 'Product_ID', x = 'Orders', data=Sales_Product_ID)

for bars in plt_10.containers:
    plt_10.bar_label(bars)
```

Hence the most demanded product is with ID P00265242
